```python
age_map = {
    '[0-10)': 5,
    '[10-20)': 15,
    '[20-30)': 25,
    '[30-40)': 35,
    '[40-50)': 45,
    '[50-60)': 55,
    '[60-70)': 65,
    '[70-80)': 75,
    '[80-90)': 85,
    '[90-100)': 95
}

df['ages'] = df['age'].map(age_map)
df['ages'].value_counts()
```

```python
df.drop('age',axis=1,inplace=True)
df
```

```python
df = df[['patient_id', 'encounter_id', 'gender', 'race','ages','days_admitted',
       'admission_type_id', 'discharge_disposition_id', 'admission_source_id',
       'lab_tests_count', 'medicine_count', 'readmitted']]
df
```

```python
df[['gender','race']]=df[['gender','race']].astype(str)
```

```python
df.dtypes
```

```python
df['treatment_intensity'] = df['lab_tests_count'] + df['medicine_count']
```

```python
df['risk_score'] = (
    df['days_admitted'] * 0.4 +
    df['lab_tests_count'] * 0.3 +
    df['medicine_count'] * 0.3
)
```

```python
def age_category(age):
    if age <= 18:
        return 'Child'
    elif age <= 35:
        return 'Young Adult'
    elif age <= 55:
        return 'Adult'
    elif age <= 75:
        return 'Middle Aged'
    else:
        return 'Senior'

df['age_category'] = df['ages'].apply(age_category)
```
